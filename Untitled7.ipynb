```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# -----------------------------
# 1. CREATE / CONNECT SQLite DB
# -----------------------------
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()

# --------------------------------------
# 2. CREATE TABLE + INSERT SAMPLE DATA
# --------------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product TEXT,
    quantity INTEGER,
    price REAL
);
""")

# Clear old data (optional â€” ensures clean reruns)
cursor.execute("DELETE FROM sales;")

sample_data = [
    ("Apple", 10, 3.0),
    ("Banana", 5, 2.0),
    ("Apple", 7, 3.0),
    ("Mango", 4, 5.0),
    ("Banana", 8, 2.0)
]

cursor.executemany("INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?);", sample_data)
conn.commit()

# -----------------------------
# 3. RUN SQL QUERY
# -----------------------------
query = """
SELECT
    product,
    SUM(quantity) AS total_qty,
    SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
ORDER BY revenue DESC;
"""

df = pd.read_sql_query(query, conn)

# -----------------------------
# 4. PRINT RESULTS
# -----------------------------
print("=== BASIC SALES SUMMARY ===")
print(df)

# -----------------------------
# 5. PLOT BAR CHART
# -----------------------------
plt.figure(figsize=(8,5))
plt.bar(df['product'], df['revenue'])
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.tight_layout()
plt.show()

# Save chart in Colab if needed
plt.savefig("sales_chart.png")

conn.close()
```
